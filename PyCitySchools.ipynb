```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#District Summary
school_count = len(school_data_complete['school_name'].unique())
student_count = len(school_data_complete['Student ID'].unique())
total_budget = '${:,.2f}'.format((school_data_complete['budget'].unique()).sum())
avg_math_score = round(school_data_complete['math_score'].mean(),1)
avg_reading_score = round(school_data_complete['reading_score'].mean(),1)

math_passes = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()
math_passing_percentage =round((math_passes/student_count)*100,2)
reading_passes = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['reading_score'].count()
reading_passing_percentage = round((reading_passes/student_count)*100,2) 

overall_rate = round((math_passing_percentage+reading_passing_percentage)/2,1) 

district_summary_df = pd.DataFrame({"Total Schools": [school_count], "Total Students":[student_count],"Total Budget": [total_budget],"Average Math Score": [avg_math_score],"Average Reading Score": [avg_reading_score],"% Passing - Math":[math_passing_percentage],"% Passing - Reading":[reading_passing_percentage],"Overall Passing Rate":[overall_rate]})
district_summary_df['% Passing - Math'] = district_summary_df['% Passing - Math'].map("{:.2f}%".format)
district_summary_df['% Passing - Reading'] = district_summary_df['% Passing - Reading'].map("{:.2f}%".format)
district_summary_df['Overall Passing Rate'] = district_summary_df['Overall Passing Rate'].map("{:.2f}%".format)
district_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#School Summary
school_info = school_data.loc[:, ['school_name', 'type', 'size', 'budget']].head(15)
school_info['Per Student Budget'] = school_data['budget'] / school_data['size']

grouped_student= student_data.groupby(['school_name'])
print(grouped_student)
grouped_student.count().head(15)

student_score = grouped_student['reading_score', 'math_score'].sum()
student_score = student_score.reset_index()

school_summary = pd.merge(student_score, school_info, on='school_name', how='outer')
school_summary['Average Reading Score'] = round(school_summary['reading_score']/ school_summary['size'],1)
school_summary['Average Math Score'] = round(school_summary['math_score']/ school_summary['size'],1)
del school_summary['math_score']
del school_summary['reading_score']

school_math_passes = student_data.loc[student_data["math_score"] >= 70,:]
passing_math_score = school_math_passes[["school_name", "math_score"]].groupby("school_name").count()
passing_math_score = passing_math_score.reset_index()
school_summary['% Passing Math'] = round((passing_math_score['math_score']/ school_summary['size'])*100,2)

school_reading_passes = student_data.loc[student_data["reading_score"] >= 70,:]
passing_reading_score = school_reading_passes[["school_name", "reading_score"]].groupby("school_name").count()
passing_reading_score = passing_reading_score.reset_index()
school_summary['% Passing Reading'] = round((passing_reading_score['reading_score']/ school_summary['size'])*100,2)

school_summary['Overall Passing Rate'] =round((school_summary['% Passing Math'] + school_summary['% Passing Reading'])/2,2)
school_summary = school_summary.rename(columns={'school_name':'School Name','type':'School Type','size':'Total Students','budget':'Total Budget'})
school_summary
```

```python
#Top Performing Schools (By Passing Rate)
top_performers = school_summary.sort_values("Overall Passing Rate", ascending = False)
top_performers.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Bottom Performing Schools (By Passing Rate)
bottom_performers = school_summary.sort_values("Overall Passing Rate")
bottom_performers.head(5)
```
